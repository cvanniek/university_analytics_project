# Bronze Layer: College Scorecard Data Ingestion

## Overview
This notebook ingests raw institutional data from the U.S. Department of Education College Scorecard dataset and prepares a structured Bronze layer dataset for downstream analytics.

The raw dataset contains 6,429 institutions and over 3,306 attributes. Many fields are administrative, redundant, or program-specific. To support analytics use cases, a subset of core institutional performance metrics is selected.

## Selected Fields
The following columns were selected to support institutional performance and affordability analysis:

| Column | Description | Reason |
|------|------|------|
| UNITID | Unique institution identifier | Primary key |
| INSTNM | Institution name | Human readable reporting |
| CITY | City | Geographic analysis |
| STABBR | State abbreviation | Regional analysis |
| CONTROL | Institution type (public/private/for-profit) | Sector comparison |
| LOCALE | Urban/suburban/rural classification | Environmental context |
| UGDS | Undergraduate enrollment | Institutional scale |
| TUITIONFEE_IN | In-state tuition | Affordability metric |
| TUITIONFEE_OUT | Out-of-state tuition | Cost comparison |
| C150_4 | Graduation rate (4 years) | Institutional performance |
| MD_EARN_WNE_P10 | Median earnings 10 years after enrollment | Outcome metric |
